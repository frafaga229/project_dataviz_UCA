```python
df_plot = df[["genres", "dbp_genre","lifeSpan.begin","lifeSpan.end", "locationInfo", "location.country"]]
df_plot.head()
```

```python
df_plot.isnull().sum()
```

```python
df_plot.isna().sum()
```

```python
for ind in df_plot.index:
    row = df_plot['locationInfo'][ind]
    n_row = row.split('"')
    if len(n_row) > 1:
        df_plot['locationInfo'][ind] = n_row[1]
    else:
        df_plot['locationInfo'][ind] = np.nan
            
df_plot.head()
```

```python
df_plot.isna().sum()
```

```python
for ind in df_plot.index:
    row = df_plot['genres'][ind]
    n_row = row.split('"')
    if len(n_row) > 1:
        df_plot['genres'][ind] = n_row[1]
    else:
        df_plot['genres'][ind] = np.nan
            
df_plot.head()
```

```python
df_plot.isna().sum()
```

```python
for ind in df_plot.index:
    
    row = df_plot['dbp_genre'][ind]
    if pd.isnull(row):
        continue
    else:
        n_row = row.split('"')
        if len(n_row) > 1:
            df_plot['dbp_genre'][ind] = n_row[1]
        else:
            df_plot['dbp_genre'][ind] = np.nan
            
#df_plot.head()
```

```python
df_plot.head()
```

```python
for ind in df_plot.index:
    row = df_plot['lifeSpan.begin'][ind]
    if pd.isnull(row):
        continue
    else:
        n_row = row.split('-')
        df_plot['lifeSpan.begin'][ind] = n_row[0]
```

```python
df_plot.head()
```

```python
df_plot = df_plot.drop(columns = ["lifeSpan.end"])
```

```python
# Since the begin year is necessary for the plot, we're going to drop all rows with nan in lifeSpan.begin colum
df_plot = df_plot.dropna(subset=['lifeSpan.begin'])
```

```python
print(df_plot.head(15))
```

```python
df_plot.shape
```

```python
df_plot.isna().sum()
```

```python
for ind in df_plot.index:
    row = df_plot['locationInfo'][ind]
    row1 = df_plot['dbp_genre'][ind]
    if pd.isnull(row):
        df_plot['locationInfo'][ind] = df_plot['location.country'][ind]
    if pd.isnull(row1):
        df_plot['dbp_genre'][ind] = df_plot['genres'][ind]
```

```python
df_plot.isna().sum()
```

```python
df_plot = df_plot.drop(columns = ["location.country", "genres"])
df_plot.head(20)
```

```python
# Keep only the rows with at least 2 non-NA values.
df_plot = df_plot.dropna(thresh=2)
df_plot.shape
```

```python
#Replacing all NaN values for 'other'
df_plot = df_plot.replace(np.nan,'Other')
df_plot.head(20)
```

```python
#replacing all rock titles for simple rock so reducing types of music
for ind in df_plot.index:
    row = df_plot['dbp_genre'][ind]
    if 'rock' in row or 'Rock' in row:
        df_plot['dbp_genre'][ind] = 'Rock'
    elif 'country' in row or 'Country' in row:
        df_plot['dbp_genre'][ind] = 'Country'
    elif 'rhythm' in row or 'Rhythm' in row:
        df_plot['dbp_genre'][ind] = 'R&B'
    elif 'pop' in row or 'Pop' in row:
        df_plot['dbp_genre'][ind] = 'Pop'
    elif 'hiphop' in row or 'hip hop' in row or 'Hip hop' in row or 'Hip Hop' in row:
        df_plot['dbp_genre'][ind] = 'Hip Hop'
    elif 'folk' in row or 'Folk' in row:
        df_plot['dbp_genre'][ind] = 'Folk'
    elif 'metal' in row or 'Metal' in row:
        df_plot['dbp_genre'][ind] = 'Metal'
    elif 'electro' in row or 'Electro' in row:
        df_plot['dbp_genre'][ind] = 'Electronic'
    elif 'blues' in row or 'Blues' in row:
        df_plot['dbp_genre'][ind] = 'Blues'
    elif 'punk' in row or 'Punk' in row:
        df_plot['dbp_genre'][ind] = 'Punk'
    elif 'reggae' in row or 'Reggae' in row:
        df_plot['dbp_genre'][ind] = 'Reggae'
    elif 'singer' in row or 'Singer' in row:
        df_plot['dbp_genre'][ind] = 'Singer-songwriter'
    elif 'contemporary' in row or 'Contemporary' in row:
        df_plot['dbp_genre'][ind] = 'Contemporary'
    elif 'jazz' in row or 'Jazz' in row:
        df_plot['dbp_genre'][ind] = 'Jazz'
    elif 'New wave' in row or 'New Wave' in row:
        df_plot['dbp_genre'][ind] = 'New Wave'
    elif 'soul' in row or 'Soul' in row:
        df_plot['dbp_genre'][ind] = 'Soul'
    elif 'hardcore' in row or 'Hardcore' in row:
        df_plot['dbp_genre'][ind] = 'Hardcore'
    elif 'gospel' in row or 'Gospel' in row:
        df_plot['dbp_genre'][ind] = 'Gospel'
    elif 'dance' in row or 'Dance' in row:
        df_plot['dbp_genre'][ind] = 'Dance'
    elif 'funk' in row or 'Funk' in row:
        df_plot['dbp_genre'][ind] = 'Funk'
    elif 'chanson' in row or 'Chanson' in row:
        df_plot['dbp_genre'][ind] = 'Chanson'
    elif 'emo' in row or 'Emo' in row:
        df_plot['dbp_genre'][ind] = 'Emo'
    elif 'House' in row or 'house' in row:
        df_plot['dbp_genre'][ind] = 'House'
    elif 'disco' in row or 'Disco' in row:
        df_plot['dbp_genre'][ind] = 'Disco'
    else:
        df_plot['dbp_genre'][ind] = 'Other'
```

```python
#replacing years for decades
for ind in df_plot.index:
    row = df_plot['year'][ind]
    if int(row) < 1960:
        df_plot['year'][ind] = 1
    elif int(row) < 1970:
        df_plot['year'][ind] = 2
    elif int(row) < 1980:
        df_plot['year'][ind] = 3
    elif int(row) < 1990:
        df_plot['year'][ind] = 4
    elif int(row) < 2000:
        df_plot['year'][ind] = 5
    else:
        df_plot['year'][ind] = 6
```
